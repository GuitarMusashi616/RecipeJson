```python
def get_item(tag, plan):
    return plan.tag_to_item[tag]
```

```python
def get_stockpile(recipe):
    sp = Stockpile()
    for item in recipe['ingredients']:
        if 'item' in item:
            sp.add(item['item'],1)
        elif 'tag' in item:
            sp.add(get_item(item['tag'], plan),1)
    return sp
```

```python
get_stockpile(recipe)
```

```python
recipe
```

```python
def recipe_into_parseable(recipe):
    sp = get_stockpile(recipe)
    arr = []
    for item, count in sp.stock.items():
        arr.append(f"{count} {item}")
    return ", ".join(arr) + f" -> {recipe['outputAmount']} {recipe['outputName']}"
```

```python
def get_resources(plan, max_item=10000):
    return ", ".join(f"{max_item} {mat}" for mat in plan.raw_materials)
```
